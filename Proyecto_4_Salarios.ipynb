```python
import pandas as pd
ruta_archivo = '/content/drive/MyDrive/Colab Data/data_af.csv'
df = pd.read_csv(ruta_archivo)
df

display(df.corr())
from matplotlib import pyplot as plt
plt.matshow(df.corr())
plt.colorbar()
plt.show()
```

```python
df.plot(kind='scatter', x='rating', y='salary', s=32, alpha=0.3)
df.plot(kind='scatter', x='age', y='salary', s=32, alpha=0.3)
df.plot(kind='scatter', x='draft_round', y='salary', s=32, alpha=0.3)
```

```python
#definir las variables x y y
X = df[['rating']] ## lleva doble corchetes por que puede ser una tabla con probablemente varias variables
y = df['salary']

display (X)
display (y)


from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)

display (X_train)
display (X_test)
display (y_train)
display (y_test)

```

```python


from sklearn.metrics import r2_score #carga el app que va a realizar el calculo de la r2
from sklearn.metrics import mean_absolute_error #

from sklearn.linear_model import LinearRegression  #carga el app que va a realizar el calculo
modelo = LinearRegression() #crea el modelo
modelo.fit(X_train, y_train) #entrena el modelo

y_test_print=modelo.predict(X_test)
r2 = r2_score(y_test, y_test_print)

print("R2:", r2)

print (modelo.intercept_)
print (modelo.coef_)
#print (r2_score)
#print (mean_squared_error)
#print (mean_absolute_percentage_error)
```
